Vamos, depois de iniciar o ambiente lendo agora a base de dados tratada, voltando a utilizar o value_counts() para passar por todas as colunas e ver com mais calma o que cada uma têm.

```python
import polars as pl
df = pl.read_csv('base_fapesp_tratada.csv')
```

```python
# Para cada coluna existente na lista de colunas de df
for col in df.columns:
    # Imprima a contagem de valores desta coluna, em ordem decrescente    
    print(df[col].value_counts(sort=True))

    # E uma linha em branco, para fiar mais organizado.
    print()
```
